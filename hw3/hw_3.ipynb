```
from lxml import etree
from collections import defaultdict
from collections import Counter
```

```
import numpy as np
from sklearn.model_selection import KFold
kf = KFold(n_splits=5)
```

```
open_corpora = etree.fromstring(open('annot.opcorpora.no_ambig_strict.xml', 'rb').read())
```

```
tags = []

for sent in open_corpora.xpath('//tokens'):
    tagged = []
    for token in sent.xpath('token'):
        word = token.xpath('@text')
        gram_info = token.xpath('tfr/v/l/g/@v')
        tagged.append(gram_info)
    
    tags.append(tagged)

tags[:3]
```

```
with open('corpus_train.txt', 'w', encoding = 'utf-8') as j:

    for sent in open_corpora.xpath('//tokens'):
        for token in sent.xpath('token'):
            word = token.xpath('@text')
            gram_info = token.xpath('tfr/v/l/g/@v')
            
            j.write(word[0] + '\t' + ','.join(gram_info) + '\n')
```

```
with open('corpus_train.txt', 'r') as m:
    corpus = m.readlines()
```

```
def train(train_set):
    corpus = open('corpus_train.txt', 'w')
    vocab = defaultdict(set)
    tags = set()
    for sentence in train_set:
        length = len(sentence.xpath('token'))
        ended = False
        for i,token in enumerate(sentence.xpath('token')):
            word = token.xpath('@text')
            gram_info = token.xpath('tfr/v/l/g/@v')
            if (i+1)==length and gram_info[0] == 'PNCT':
                 gram_info = ['SENT']
                 ended = True
            corpus.write(word[0] + '\t' + ','.join(gram_info) + '\n')
            lemma = token.xpath('tfr/v/l/@t')[0]
            vocab[word[0].lower()].add((','.join(gram_info), lemma.lower()))
            tags.add(','.join(gram_info))   
        if not ended:
            corpus.write('.\tSENT\n')

    with open('lexicon.txt', 'w') as f:
        for word in vocab:
            f.write(word + '\t')
            f.write('\t'.join([' '.join(pair) for pair in vocab[word]]))
            f.write('\n')
    with open('open_class.txt', 'w') as t:
        t.write('\n'.join([tag for tag in tags if 'NOUN' in tag or 'VERB' in tag or 'ADJF' in tag]))      
    !./bin/train-tree-tagger lexicon.txt open_class.txt corpus_train.txt model_oc
    return
    
```

```
def testing(sents):
  words = []
  for sentence in open_corpora.xpath('//tokens'):
    tokens = sentence.xpath('token')
    for token in tokens:
      word = token.xpath('@text')
      words.append(str(word))

  test = open('test_corpus.txt', 'w')
  test.write('\n'.join(words))
  test.close()

  !./bin/tree-tagger model_oc test_corpus.txt output.txt
  
  with open('output.txt', 'r', encoding = 'utf-8') as j:
    res = j.read()
    res.split('\n')
    return res
```

```
def jaccard(check, res):
  
  preds = []
  mistakes = Counter()

  for tags, prediction in zip(check, res):
    pred = set(prediction.split(','))
    tag = set([str(i) for i in tags])
    p = len(pred&tag)/len(pred|tag)
    preds.append(p)
    
  return preds
```

```
jacs = []

for train, test in kf.split(np.array(sents)):
  train_set = [sents[i] for i in train]
  test_set = [sents[i] for i in test]
  result = testing(test_set)
  
  check = [tags[i] for i in test]
  jacs.append(jaccard(check, result))

jacs
```

```
sents = open_corpora.xpath('//tokens')

for train_i, test_i in kf.split(np.array(sents)):
    train_set = [sents[i] for i in train_i]
    test_set = [sents[i] for i in test_i]
    train(train_set)
    !./bin/tree-tagger model_oc test_corpus.txt output.txt
```
